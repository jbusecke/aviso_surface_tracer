# Full Variance Budget for Busecke and Abernathey 2017

In this notebook I evaluate the full tracer variance budget for all runs assembled for the 2017 ENSO diffusivity paper.

The full budget (from Abernathey and Marshall 2013) can be written as:

$$ 
\underbrace{\frac{\partial}{\partial t}\frac{\overline{{q^\prime}^2}}{2}}
_\text{tendency} + 
\underbrace{\nabla \cdot \overline{v \frac{{q^\prime}^2}{2}}}
_\text{advection}+ 
\underbrace{\overline{v^\prime q^\prime} \cdot \nabla \overline{q}}
_\text{production}=
\underbrace{\nabla^2 \left({\kappa \overline{ \frac{{q^\prime}^2}{2}}}\right)}
_\text{diffusion} - 
\underbrace{\kappa \overline{|{\nabla q^\prime}|^2}}
_\text{destruction}$$

I do have the following 

% For the paper: \usepackage{amsmath} 

The following terms are put out as diagnostics from the model:

- $$\overline{q}$$
- $$\overline{q^2}$$
- $$

The terms are calculated as follows:

Production: 
$$
\overline{v^\prime q^\prime} \cdot \nabla \overline{q} =
\overline{u^\prime q^\prime} \frac{\partial \overline{q}}{\partial x} +
\overline{v^\prime q^\prime} \frac{\partial \overline{q}}{\partial y} = 
(\overline{uq} - \overline{u} \, \overline{q}) \frac{\partial \overline{q}}{\partial x} +
(\overline{vq} - \overline{v} \, \overline{q}) \frac{\partial \overline{q}}{\partial y}
$$

Advection:
$$
\nabla = \overline{ \vec{u} \frac{{q^\prime}^2}{2}} = 
\frac{\partial}{\partial x} \overline{u \frac{{q^\prime}^2}{2}}
$$

```python
# Basic vector algebra (This should update the aviso tracer repo)
def _get_matching(da, da_list, debug=False):
    """from a list of xarray data arrays (da_list) return the one with dims 
    matching obj"""
    matched = [a for a in da_list if set(a.dims).issubset(set(da.dims))]
    if debug:
        print(matched)
    if len(matched) == 0:
        raise RuntimeError('no matching coordinates found. Try interpolating them using xgcm')
    elif len(matched) >1 :
        raise RuntimeError('More than one matching coord found, check list of data arrays')
    else:
        matched = matched[0]
    return matched

def _get_dx(da, grid_ds, debug=False):
    """get horizontal distance"""
    dx_list = [grid_ds[a] for a in ['dxC', 'dxG']]
    return _get_matching(da,dx_list, debug=debug)

def _get_dy(da, grid_ds, debug=False):
    """get horizontal distance"""
    dy_list = [grid_ds[a] for a in ['dyC', 'dyG']]
    return _get_matching(da,dy_list, debug=debug)

def _get_hfac(da, grid_ds, debug=False):
    """get horizontal distance"""
    hfac_list = [grid_ds[a] for a in ['hFacC', 'hFacW', 'hFacS']]
    return _get_matching(da,hfac_list, debug=debug)
    
def _get_area(da, grid_ds, debug=False):
    area_list = [grid_ds[a] for a in ['rA', 'rAw', 'rAs']]
    return _get_matching(da,area_list, debug=debug)

def gradient(da, grid):
    da = da.copy()    
    delta_x = grid.diff(da, 'X')
    delta_y = grid.diff(da, 'Y')
    dx = _get_dx(delta_x,grid._ds)
    dy = _get_dy(delta_y,grid._ds)
    da_dx = delta_x/dx
    da_dy = delta_y/dy
    return da_dx, da_dy

# def dot_prod(u1, v1, u2, v2, grid, debug=False):
#     u = u1*u2
#     v = v1*v2
#     if debug:
#         print(u)
#         print(v)
    
#     dyu = _get_dy(u, grid._ds, debug=debug)
#     dxv = _get_dx(v, grid._ds)
#     hfacu = _get_hfac(u, grid._ds)
#     hfacv = _get_hfac(v, grid._ds)
#     drF = grid._ds['drF']
    
#     u_int = grid.interp(u*dyu*hfacu*drF, 'X')
#     v_int = grid.interp(v*dxv*hfacv*drF, 'Y')
    
#     dot_prod_int = u_int+v_int
    
#     area = _get_area(dot_prod_int, grid._ds)
#     hfac = _get_hfac(dot_prod_int, grid._ds)
#     return dot_prod_int/(area*hfac*drF)

def dot_prod(u1, v1, u2, v2, grid, debug=False):
    u1 = u1.copy()
    u2 = u2.copy()
    v1 = v1.copy()
    v2 = v2.copy()    
    u = u1*u2
    v = v1*v2
    
    u_int = grid.interp(u, 'X')
    v_int = grid.interp(v, 'Y')
    
    dot_prod = u_int+v_int
    return dot_prod

def divergence(u, v, grid):
    u = u.copy()
    v = v.copy()
    dyu = _get_dy(u, grid._ds)
    dxv = _get_dx(v, grid._ds)
    hfacu = _get_hfac(u, grid._ds)
    hfacv = _get_hfac(v, grid._ds)
    drF = grid._ds['drF']
    transport_u = u*dyu*hfacu*drF
    transport_v = v*dxv*hfacv*drF
    delta_transport_u = grid.diff(transport_u,'X')
    delta_transport_v = grid.diff(transport_v,'Y')
    
    area = _get_area(delta_transport_u, grid._ds)
    hfac = _get_hfac(delta_transport_u, grid._ds)
    # Volume integrated divergence
    div_uv = (delta_transport_u+delta_transport_v)/(area*hfac*drF)
    return div_uv

def laplacian(da, grid):
    da = da.copy()
    gradx, grady = gradient(da, grid)
    return divergence(gradx, grady, grid)
```

```python
# This could potentially be integrated in xmitgcm
def convert_trnum2dimension(ds):
    def drop_nonmatching_vars(ds, var_list):
        data_vars = list(ds.data_vars)
        drop_vars = [a for a in data_vars if a not in var_list]
        return ds.drop(drop_vars)
    
    def rename_vars(ds):
        for vv in list(ds.data_vars):
            ds = ds.rename({vv:vv[0:-2]})
        return ds
        
    tr_num = list(set([a[-2:] for a in list(ds.data_vars)]))
    tr_num.sort()
    tr_vars = list(set([a[:-2] for a in list(ds.data_vars)]))
    
    datasets = [drop_nonmatching_vars(ds, [a+n for a in tr_vars]) for n in tr_num]
    datasets = [rename_vars(b) for b in datasets]
    
    tr_num_int = [int(a) for a in tr_num]
    
    tr_dim = xr.DataArray(tr_num_int, 
                          dims='tracer_no',
                          coords={'tracer_no': (['tracer_no', ], tr_num_int)})
    
    ds_combined = xr.concat(datasets,dim='tracer_no')
    ds_combined['tracer_no'] = tr_dim
    
    return ds_combined

# These should all go to the final version of the github repo
```

# Exercise 1: Infer global numerical diffusivity from globally integrated variance budget
The globally integrated variance budget is given by:
$$ \frac{\partial}{\partial t} \iint \frac{q^2}{2} dA = -\kappa_{num} \iint |\nabla q^2| \,dA$$

We can estimate the tendency term on the LHS using the snapshot diagnostics and the gradient variance term on thr RHS using the mean diagnostics, and solve for $\kappa_{num}$. This is only possible under the assumption that $\kappa_{num}$ is spatially constant.

```python
# Determine the numerical diffusivity from globally integrated variance budget
inferred_kappa = infer_kappa_from_global_variance_budget(ds_snap, ds_mean, grid).compute()
```

# Exercise 2: Mean Tracer budget
After Reynolds decomposition of the tracer field, the mean tracer budget is given by:
$$ \frac{\partial}{\partial t} \overline{q} = 
-\vec{u} \cdot \nabla \overline{q} 
-\nabla \cdot \overline{\vec{u}^\prime q^\prime}
+\kappa \nabla^2 \overline{q}$$

```python
q_snap = ds_snap['TRAC']
q_mean = ds_mean['TRAC']
u_pr_q_pr = ds_mean['UTRAC'] - (ds_vel['UVEL']*grid.interp(q_mean, 'X'))
v_pr_q_pr = ds_mean['VTRAC'] - (ds_vel['VVEL']*grid.interp(q_mean, 'Y'))

dt = ds_snap['TRAC'].time[0:2].diff('time').data

simple_tendency = q_snap.diff('time')/dt
simple_gradx_q, simple_grady_q = gradient(q_mean, grid)

simple_advection_mean = - dot_prod(simple_gradx_q, simple_grady_q, 
                                   ds_vel['UVEL'], ds_vel['VVEL'], 
                                   grid, q_mean)
simple_advection_eddy = - divergence(u_pr_q_pr, v_pr_q_pr, grid)
simple_diffusion = kappa*laplacian(q_mean, grid)

simple_sum = simple_advection_mean+simple_diffusion+simple_advection_eddy

# quick order of magnitude sanity check.
simple_combo = xr.concat([simple_tendency.drop(simple_tendency.coords),
                       simple_diffusion.drop(simple_diffusion.coords),
                       simple_advection_mean.drop(simple_advection_mean.coords),
                       simple_advection_eddy.drop(simple_advection_eddy.coords),
                       simple_sum.drop(simple_sum.coords)],
                dim=['tendency',
                     'diffusion',
                     'adv_mean',
                     'adv_eddy',
                     'sum'])
simple_combo
```
